# Read Data From JSON File

```python
import json
file=open('record.json','r')
r=file.read()
file.close()

record=json.loads(r)
```

```python
record
```

# Adding New Item

```python
choice=True

while(choice):
    product_id=int(input("Enter the Product ID: "))
    if(str(product_id) in record.keys()):
        print(record[str(product_id)])
    prod_name=input("Enter the name of the product: ")
    price=int(input('Enter the Price: '))
    quantity=int(input('Enter the quantity: '))
    mfd=input("Enter the Date of Manufacture Date: ")
    exd=input("Enter the Expiry Date: ")

    if(str(product_id) in record.keys()):
        

        record[str(product_id)]['quantity']+=quantity
        record[str(product_id)]['mfd']=mfd
        record[str(product_id)]['exd']=exd
        print("******************************************************************************")
        print("After Updation: ")
        print(record[str(product_id)])
        print("******************************************************************************")
        
    else:
        record[product_id]={'name':prod_name,'price':price,'quantity':quantity,'mfd':mfd,'exd':exd}
        
    print("Do you want to add any new product? ")
    print("Type Y/y for YES and N/n for NO ")
    ch=input("Enter Y/y or N/n: ")
    if(ch=="Y" or ch=="y"):
        choice=True
    elif(ch=="N" or ch=="n"):
        choice=False
```

# Updating Inventory after addition

```python
js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
